### Mixed Variable:

- When in any coulmn data is in form of alphanumeric like B25, C5 and so on then we say that that particular column has mixed data of it is mixed variable or attribute. This type of variable we mostly see in our real life when we come accross railway ticket, air ticket or any kind of ticket number.



- When we get this type of data, the most important thing we need to do is that we need to separate the numerical part and categorical (alphabet) part.



- Also there is a one more form of mixed variable that is some record will have numeric value and some record will have categorical value. So in this type of variable we create two separate column one with numeric data and other with categorical data. And in numeric column we fill NAN where data is categorical and in categorical coulmn we fill NAN where data is numeric.





- We are going to see both type of mixed variable in this notebook:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

- Here in Ticket and Cabin column we can see that data is present in same form as i mentioned above.

#### number column:

- Here number column is showing how many family members were travelling with a particular person.
- In this column you will notice that some record has value "A" which means that particular person was travelling Alone.
- So this is example of second type of mixed variable.

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with number of family members')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

### Cabin and ticket column:

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

- Both cabin and ticket have lots of unique values

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

- After extracting category, we have less category to deal with

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
