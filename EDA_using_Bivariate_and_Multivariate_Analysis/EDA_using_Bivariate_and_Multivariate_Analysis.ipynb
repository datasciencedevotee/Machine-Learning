## Exploratory Data Analysis: Bivariate, and Multivariate Analysis

- Analsyis can be done by two ways:
  - Statistical Method
  - Graphical Method
  
- Here we will be mostly discussing graphical methods and try to look some of the statistical approach.

### What is Bivariate Analysis?

- Bi means two and variate means variable. Collectively, Bivariate analysis refers to the exploratory data analysis between two variables. The variables can be either numeric or categorical. Bivariate analysis helps in studying the relationship between two variables.

### What are the types of bivariate analysis?
#### We know the types of data can be either numerical or categorical. So there can be three types of scenarios:

- Numerical feature vs. Numerical feature
- Categorical feature vs. Categorical feature
- Numerical feature vs. Categorical features

### What is Multivariate Analysis?

- 'Multi' means many, and 'variate' means variable. Multivariate analysis refers to the statistical procedure for analyzing the data involving more than two variables. 


- Alternatively, this can be used to analyze the relationship between dependent and independent variables. 


- Multivariate analysis has various applications in clustering, feature selection, root-cause analysis, hypothesis testing, dimensionality reduction, etc.

### 1. Scatterplot (Numerical vs. Numerical)

- A scattter plot or scatter graph is used to plot data points corresponding to two features. This helps explain the change in one variable with respect to the change in the other one. A dot in the scatterplot represents each row of the dataset. This also helps explian the correlation between two variables, but primarily, scatter plots are used to establish the relationship between two variables.

```python
sns.scatterplot(data=iris, x='sepal_length', y="petal_length", hue='species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

- The above scatterplot clearly shows the presence of three distinct clusters of different flower species. On the X-axis, we have the Sepal length of the flower, while on the Y-axis, we have the Petal Length. The scatterplot indicates a strong positive correlation between sepal Length and Petal Length.

#### How can we comment on the correlation just by looking at the scatterplot?

- The image below will illustrate how we can comment on the correlation between two variables by looking at the scatterplot.

- Correlation varies between -1 to 1. 
- A correlation of positive one indicates a perfect positive linear relationship.
- While anegative one indicates a perfectly inverse relationship between two variables.
- Further, a correlation of zero indicates no connection between the two variables.

```python
## Scatterplot between total bill and tips

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.xlabel("Total Bill paid")
plt.ylabel("Tip paid")
plt.show()
```

- The above scatterplot indicates a positive correlation between Total_Bill and Tip.

```python
## Scatterplot between total bill and tips and 
## use hue parameter to separate it according to sex

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.xlabel("Total Bill paid")
plt.ylabel("Tip paid")
plt.show()
```

The above scatterplot clearly shows the presence of two distinct clusters of male and female. On the X-axis, we have total bill paid by people, while on the Y-axis, we have tip paid by people. The scatterplot indicates a positive correlation between total bill and tip.

### 3. Boxplot (Numerical vs. Categorical)

- A boxplot or whisker plot is a diagram often used for visualizing the distribution of numeric values. We can also use boxplot to see distribution of numeric data according different categories.

- Boxplot is also very useful to find out outliers in numeric data.


- A boxplot divides the data into equal parts using three quartiles, and this serves as an excellent visualization  of distribution.


- A boxplot consists of the lowest value, the first quartile (Lower Quartile), the second quartile (Median), the Third quartile (Upper Quartile), and finally, the highest value. A quartile is a statistical term used to describe the division of observations. The mentioned three quartiles divide the data into four equal parts. This can be confirmed using the illustration given below:

```python
import numpy as np

x = np.random.normal(0, 1, 10000)
mean = x.mean()
std = x.std()
q1, median, q3 = np.percentile(x, [25, 50, 75])
iqr = q3 - q1

plt.figure(figsize=(16,4))

medianprops = dict(linestyle='-', linewidth=2, color='yellow')
sns.boxplot(x=x, color='#009ACD', saturation=1, medianprops=medianprops,
             flierprops={'markerfacecolor': 'mediumseagreen'}, whis=1.5)
```

- The above box plot is generated from a normal distribution, and because of that, it is approximately symmetric with respect to the middle yellow line.

- The **Inter Quartile Range** (IQR) represents the middle 50% values. Each quartile to end or quartile covers 25% of the data. Hence, IOR is the difference between the third and the first quartile.


- **IOR = Q3 - Q1**, IQR can be used to find the outliers in the data.


- Boxplot can help in visualizing the distribution of data. The image below can distinguih the skewed distribution vs. the normal distribution pattern.

```python
## Boxplot of Age (Numeric data) categories by Sex (Categorical data)

sns.boxplot(data = titanic, x="Sex", y="Age")
plt.show()
```

- Almost same distribution of age in both category.
- Median age is same in both male and female.
- In male we can see that some outliers are present.

```python
## Boxplot with hue parameter

sns.boxplot(data = titanic, x="Sex", y="Age", hue="Survived")
plt.show()
```

### 4. Distplot (Numerical vs. Categorical)

- In titanic dataset, basically we have target feature which is survived.

- Lets filter data according to survived and not survived.

- And then let see distribution of age in both cases with KDE (Probability)

```python
not_survived = titanic[titanic["Survived"]==0]
survived = titanic[titanic["Survived"]==1]
```

```python
sns.distplot(not_survived["Age"], hist = False) # probability of dying
sns.distplot(survived["Age"], hist = False) # probability of surviving
plt.show()
```

- Probability of surviving for age group 0 to 15 is more than probability of dying.


- Probability of surviving and dying is varying according to age group.

### 5. HeatMap (Categorical vs. Categorical)

- A heatmap (or heat map) is a graphical representation of data where values are depicted by color.

```python
titanic.sample(2)
```

```python
## will make cross tab of passenger class and survived attributes

cross_tab = pd.crosstab(titanic["Pclass"], titanic["Survived"])
cross_tab
```

```python
## will visualize that cross tab using heat map

sns.heatmap(cross_tab)
plt.show()
```

- According to cross tab or heat map it clear that more people survived in pclass one but it may possible that it is because that more passengers were travelling in pclass one.


- So lets find percentage of survival according to passenger class.


- For finding percenatge, we will use groupby method in pandas. We will group by data according to passenger class and then find percentage of survival in each passenger class.

#### Percentage of survival in different passenger class:

```python
titanic.groupby("Pclass").mean()["Survived"]*100
```

- Survival percentage is more in pasenger class one.

#### Percentage of survival in different gender:

```python
titanic.groupby("Sex").mean()["Survived"]*100
```

### 8. Lineplot (Numerical vs Numerical)

- A line plot can be defined as a graph that displays data as points or check marks above a number line, showing the frequency of each value.


- It is very useful for time related variabel where we want to see a trend with time. 


- In X-axis we keep time variable and in Y-axis we put frequency.

```python
flights.sample(2)
```

```python
## grouping data with year and then performing sum operation
## new data frame created
passenger_per_year = flights.groupby("year").sum().reset_index()
passenger_per_year.sample(4)
```

```python
## Line plot to see trend
sns.lineplot(passenger_per_year["year"], passenger_per_year["passengers"])
```

```python
## Pivot table: Can get number of passenger travelling in a particular month in a year

passengers_per_month = flights.pivot_table(values="passengers", index="month", columns="year")
passengers_per_month.sample(4)
```

- With this pivot table it is not easy to find any trend in data.


- If we plot this data frame with some colour scheme chart than it will become easy to find some trend if it is available.

```python
## With Heat map it is easy to see some trend if available

plt.figure(figsize=(8,8))
sns.heatmap(passengers_per_month)
plt.show()
```

- This heat map shows us that between 1956 to 1960 in month of may to aug, passenger count is more.
